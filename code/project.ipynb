```python
import cv2
from ultralytics import YOLO

# Load your custom YOLO model (replace with the actual path to your model)
model = YOLO("/Users/samradny/.pyenv/runs/detect/yolov8n_custom_pholoe4/weights/best.pt")

# URL for Elgato Camera Hub (replace with actual camera feed URL)
camera_url = 'http://192.168.1.225:8080/video'  # Replace x with each of the IP addresses
  # Replace with your Elgato camera IP

# Open a connection to the camera feed
cap = cv2.VideoCapture(camera_url)

if not cap.isOpened():
    print("Error: Unable to open camera feed.")
    exit()

while True:
    # Read frames from the camera
    ret, frame = cap.read()

    if not ret:
        print("Error: Failed to capture image.")
        break

    # Perform YOLO detection on the frame
    results = model(frame)

    # Loop through detections and draw bounding boxes for potholes
    for result in results:
        for box in result.boxes:
            # Get the label for the detection (e.g., 'pothole' in your case)
            label = result.names[int(box.cls)]

            # If the label is 'pothole', process the detection
            if label == 'pothole':
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # Extract bounding box coordinates
                confidence = box.conf  # Confidence score

                # Draw the bounding box and label on the frame
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, f"{label} {confidence:.2f}", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display the frame with detections
    cv2.imshow('YOLO Pothole Detection', frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()


```
